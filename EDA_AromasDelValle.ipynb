--------- K P I ---------

```python
# Configuración visual
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Crear columna de valor total del inventario
df['InventoryValue'] = df['Price'] * df['onHand']

# Agrupar por ciudad y sumar inventario
df_city_inventory = df.groupby('City', as_index=False)['onHand'].sum()

# Ordenar de mayor a menor inventario
top_city_inventory_sorted = df_city_inventory.sort_values(by='onHand', ascending=False)

# Mostrar la ciudad con mayor inventario
top_city_inventory = df_city_inventory_sorted.head(1)

# Agrupar por ciudad y contar tiendas únicas
top_city_store = df.groupby("City")["Store"].nunique().sort_values(ascending=False).head(10)

# ==========================
# 1️⃣ KPI PRINCIPALES
# ==========================
print("📊 KPI PRINCIPALES")
print(f"- Total de productos únicos: {df['Description'].nunique()}")
print(f"- Total de marcas únicas: {df['Brand'].nunique()}")
print(f"- Valor total del inventario: ${df['InventoryValue'].sum():,.2f}")
print(f"- Precio promedio de productos: ${df['Price'].mean():.2f}")
print(f"- Ciudad con mayor inventario: {top_city_inventory['City'].iloc[0]} ({top_city_inventory['onHand'].iloc[0]:,.0f} units)")
print(f"- Ciudad con mas tiendas: {top_city_store.index[0]} ({top_city_store.iloc[0]:,.0f} units)")

# ==========================
# 2️⃣ TOP 10 POR City vs onHand
# ==========================

plt.figure(figsize=(10,6))
sns.barplot(x='onHand', y='City', data=top_city_inventory_sorted.head(10), palette="viridis", hue='City', legend=False)
plt.title("Top 10 Ciudades con Mayor Inventario", fontsize=14, weight='bold')
plt.xlabel("Inventario Total (onHand)")
plt.ylabel("Ciudad")

# Agregar etiquetas con valores
for index, value in enumerate(top_city_inventory_sorted.head(10)['onHand']):
    plt.text(value + 5, index, f"{value:,}", va='center')

plt.tight_layout()
plt.show()


# ==========================
# 2️⃣ TOP 5 City vs onHand
# ==========================

# Agrupar por ciudad: sumar inventario y valor
df_city = df.groupby('City', as_index=False).agg({
    'onHand': 'sum',
    'InventoryValue': 'sum'
})

# Ordenar por inventario y tomar las 5 ciudades principales
top5_city = df_city.sort_values(by='onHand', ascending=False).head(5)

# Crear gráfico combinado
fig, ax1 = plt.subplots(figsize=(12,6))

# Barras: inventario
sns.barplot(x='City', y='onHand', data=top5_city, palette="viridis", ax=ax1,hue='City', legend=False)
ax1.set_xlabel("Ciudad")
ax1.set_ylabel("Inventario Total (onHand)", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
plt.title("Top 5 Ciudades: Inventario vs Valor Total de Inventario", fontsize=14, weight='bold')

# Etiquetas en las barras de inventario
for i, value in enumerate(top5_city['onHand']):
    ax1.text(i, value + 50, f"{value:,}", ha='center', color="blue", fontweight='bold')

# Línea: valor total de inventario
ax2 = ax1.twinx()
ax2.plot(top5_city['City'], top5_city['InventoryValue'], color="red", marker="o", linewidth=2, label="Valor Total ($)")
ax2.set_ylabel("Valor Total de Inventario ($)", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# Etiquetas en los puntos de valor
for i, value in enumerate(top5_city['InventoryValue']):
    ax2.text(i, value + (value * 0.02), f"${value:,.0f}", ha='center', color="red", fontweight='bold')

plt.tight_layout()
plt.show()

# ==========================
# 2️⃣ TOP 10 Size vs OnHand
# ==========================

# Agrupar por tamaño (Size) y sumar inventario (onHand)
top10_size_inventory = df.groupby('Size', as_index=False)['onHand'].sum()
top10_size_inventory = top10_size_inventory.sort_values(by='onHand', ascending=False).head(10)

# Gráfico de barras vertical
plt.figure(figsize=(12,6))
sns.barplot(x='Size', y='onHand', data=top10_size_inventory, palette='Blues_r', hue='Size', legend=False)
plt.title("Top 10 Tamaños (Size) por Cantidad de Inventario")
plt.xlabel("Tamaño (Size)")
plt.ylabel("Inventario (OnHand)")

# Agregar etiquetas encima de cada barra
for index, value in enumerate(top10_size_inventory['onHand']):
    plt.text(index, value + (value * 0.01), f"{int(value):,}", ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# ==========================
# 3️⃣ TOP 10 POR SUMA DE PRICE
# ==========================
df_sum_price = df.groupby('Description', as_index=False)['Price'].sum()
df_top10_price = df_sum_price.sort_values(by='Price', ascending=False).head(10)

plt.figure()
sns.barplot(x='Price', y='Description', data=df_top10_price,hue='Description', palette="viridis", legend=False)
plt.title("Top 10 Productos por Suma de Price")
plt.xlabel("Suma de Price")
plt.ylabel("Descripción")

# Agregar etiquetas con valores
for index, value in enumerate(df_top10_price['Price']):
    plt.text(value + 5, index, f"{value:,}", va='center')

plt.tight_layout()   # 🔥 Ajusta márgenes automáticamente
plt.show()

# ==========================
# 4️⃣ DISTRIBUCIÓN DE PRECIOS
# ==========================
plt.figure()
sns.histplot(df['Price'], bins=20, kde=True, color="blue")
plt.title("Distribución de Precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")

plt.tight_layout()   # 🔥 Ajusta márgenes automáticamente
plt.show()

# ==========================
# 5️⃣ VALOR DE INVENTARIO POR CIUDAD
# ==========================
df_city_value = df.groupby('City', as_index=False)['InventoryValue'].sum()
plt.figure()
sns.barplot(x='InventoryValue', y='City', data=df_city_value.sort_values('InventoryValue', ascending=False),hue='City', palette="magma", legend=False)
plt.title("Valor de Inventario por Ciudad")
plt.xlabel("Valor de Inventario ($)")
plt.ylabel("Ciudad")

plt.tight_layout()   # 🔥 Ajusta márgenes automáticamente
plt.show()

# ==========================
# 6️⃣ PARTICIPACIÓN DE MARCAS (TOP 10)
# ==========================
df_brand_value = df.groupby('Brand', as_index=False)['InventoryValue'].sum()
df_brand_top10 = df_brand_value.sort_values(by='InventoryValue', ascending=False).head(10)

plt.figure()
plt.pie(df_brand_top10['InventoryValue'], labels=df_brand_top10['Brand'], autopct='%1.1f%%', startangle=140)
plt.title("Participación de Marcas en Inventario (Top 10)")

plt.tight_layout()   # 🔥 Ajusta márgenes automáticamente
plt.show()

# ==========================
# 7️⃣ PRODUCTOS CON MAYOR STOCK Y VALOR
# ==========================
df_top_stock = df.sort_values(by='onHand', ascending=False).head(10)
plt.figure()
sns.barplot(x='onHand', y='Description', data=df_top_stock, palette="coolwarm",hue='Description', legend=False)
plt.title("Top 10 Productos con Mayor Stock (onHand)")
plt.xlabel("Cantidad en Stock")
plt.ylabel("Descripción")

# Agregar etiquetas con valores
for index, value in enumerate(df_top_stock['onHand']):
    plt.text(value + 5, index, f"{value:,}", va='center')

plt.tight_layout()   # 🔥 Ajusta márgenes automáticamente
plt.show()

# === Top 10 Productos con Mayor Valor en Inventario ===

# Calcular el valor de inventario por producto (Description)
top10_value = df.groupby('Description', as_index=False)['InventoryValue'].sum()
top10_value = top10_value.sort_values(by='InventoryValue', ascending=False).head(10)


# Gráfico de barras
plt.figure(figsize=(12,6))
sns.barplot(x='InventoryValue', y='Description', data=top10_value, palette='crest', hue='Description', legend=False)
plt.title("Top 10 Productos con Mayor Valor en Inventario")
plt.xlabel("Valor de Inventario ($)")
plt.ylabel("Producto")

# Agregar etiquetas con los valores en cada barra
for index, value in enumerate(top10_value['InventoryValue']):
    plt.text(value + (value * 0.01), index, f"${value:,.0f}", va='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# === Top 10 Tamaños por Precio Promedio con etiquetas dentro ===

# Calcular el precio promedio por tamaño (Size)
top10_size_price = df.groupby('Size', as_index=False)['Price'].mean()
top10_size_price = top10_size_price.sort_values(by='Price', ascending=False).head(10)

# === Gráfico de barras vertical ===
plt.figure(figsize=(14,7))
sns.barplot(x='Size', y='Price', data=top10_size_price, palette='coolwarm', hue='Size', legend=False)
plt.title("Top 10 Tamaños (Size) por Precio Promedio", fontsize=14, fontweight='bold')
plt.xlabel("Tamaño (Size)", fontsize=12)
plt.ylabel("Precio Promedio ($)", fontsize=12)

# Agregar etiquetas con los valores en cada barra
for index, value in enumerate(top10_size_price['Price']):
    plt.text(index, value + (value * 0.01), f"${value:,.0f}", ha='center', fontsize=9, fontweight='bold')


plt.tight_layout()
plt.show()



```

RESUMEN GERENCIAL

```python
# Configuración visual
sns.set(style="whitegrid")
plt.rcParams["axes.labelweight"] = "bold"

df['InventoryValue'] = df['Price'] * df['onHand']

# ==========================================================
# 🔑 ANÁLISIS GENERAL DINÁMICO
# ==========================================================

# Top 5 ciudades por inventario
city_inventory = df.groupby('City', as_index=False)['onHand'].sum().sort_values(by='onHand', ascending=False)
top_cities = city_inventory['City'].head(5).tolist()

# Ordenar de mayor a menor inventario
top_city_inventory_sorted = df_city_inventory.sort_values(by='onHand', ascending=False)

# Mostrar la ciudad con mayor inventario
top_city_inventory = df_city_inventory_sorted.head(1)

# Producto con mayor inventario
top_product_inventory = df.groupby('Description', as_index=False)['onHand'].sum().sort_values(by='onHand', ascending=False).iloc[0]

# Size con mayor inventario
size_inventory = df.groupby('Size', as_index=False)['onHand'].sum().sort_values(by='onHand', ascending=False)
top_size = size_inventory.iloc[0]

# Producto más costoso nacional
max_price_product = df.loc[df['Price'].idxmax()]

# Producto más costoso por ciudad
city_max_price = df.loc[df.groupby('City')['Price'].idxmax(), ['City','Description','Price']]
top10_city_max_price = city_max_price.sort_values(by='Price', ascending=False).head(10)

# Top 10 marcas por valor de inventario
brand_value = df.groupby('Brand', as_index=False)['InventoryValue'].sum().sort_values(by='InventoryValue', ascending=False).head(10)

# ==========================================================
# 📊 GRÁFICOS DINÁMICOS
# ==========================================================

## Top 10 Ciudades por Inventario
plt.figure(figsize=(12,6))
sns.barplot(x='onHand', y='City', data=city_inventory.head(10), palette='viridis', hue='City', legend=False)
plt.title("Top 10 Ciudades por Inventario Total")
plt.xlabel("Inventario (onHand)")
plt.ylabel("Ciudad")

# Agregar etiquetas con valores
for index, value in enumerate(brand_value.head(10)['InventoryValue']):
    plt.text(value + 1, index, f"{value:,}", va='center')

plt.tight_layout()
plt.show()

## Top 10 Marcas por Valor de Inventario
plt.figure(figsize=(12,6))
sns.barplot(x='InventoryValue', y='Brand', data=brand_value, palette='cubehelix',hue='Brand', legend=False)
plt.title("Top 10 Marcas por Valor de Inventario")
plt.xlabel("Valor de Inventario ($)")
plt.ylabel("Marca")
plt.xticks(rotation=45) # Es una función que se utiliza para rotar las etiquetas 45 grados.
plt.tight_layout()
plt.show()

## Producto más Costoso por Ciudad (Top 10)
plt.figure(figsize=(14,7))
sns.barplot(x='Price', y='City', data=top10_city_max_price, palette='magma', hue='City', legend=False)
plt.title("Top 10 Ciudades - Producto más Costoso")
plt.xlabel("Precio Máximo ($)")
plt.ylabel("Ciudad")

for index, row in enumerate(top10_city_max_price.itertuples()):
    plt.text(row.Price + (row.Price*0.01), index, f"{row.Description} (${row.Price:,.2f})", va='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# ==========================================================
# 📝 RESUMEN GERENCIAL EN MARKDOWN (DINÁMICO)
# ==========================================================

resumen = f"""
# 📊 Resumen Gerencial de Inventario y Precios

## 🔑 Análisis General
- **Ciudad líder en inventario:** {top_cities[0]}.
- **Top 5 ciudades por inventario:** {', '.join(top_cities)}.
- **Producto con mayor inventario:** {top_product_inventory['Description']} ({int(top_product_inventory['onHand'])} unidades).
- **Tamaño (Size) dominante:** {top_size['Size']} ({int(top_size['onHand'])} unidades).
- **Producto más costoso nacional:** {max_price_product['Description']} (Precio: ${max_price_product['Price']:,.2f}).

## 🏙 Ciudades
- **El inventario está concentrado en:** {top_city_inventory['City']} ({int(top_product_inventory['onHand'])} unidades).
- Producto premium destacado por ciudad: basado en el análisis dinámico de precios máximos.

## 🏷 Marcas
- **Top marcas por valor de inventario:** {', '.join(brand_value['Brand'].head(5).astype(str).tolist())}.

## 🚀 Conclusiones
✅ El inventario está centralizado en pocas ciudades principales.
✅ El tamaño **{top_size['Size']}** domina tanto en volumen como en valor económico.
✅ Productos premium como **{max_price_product['Description']}** son clave para segmentación de alto valor.
✅ Marcas líderes concentran la mayor parte del valor de inventario, relevantes para negociación comercial.

## 🎯 Próximos pasos sugeridos
- Analizar rotación temporal de inventarios.
- Relacionar ventas con stock para proyecciones.
- Automatizar dashboards interactivos para gerencia (Plotly, Power BI).
"""

display(Markdown(resumen))
```
